```python


## ➡️ PART B — Move to a *later* notebook (Sampling, ADC, digitizers)

This material is excellent, but it belongs **after** students learn:

* sampling
* Nyquist
* aliasing
* decimation

I would move and slightly rewrite it as the **opening or mid-section** of your future ADC/digitizer notebook.

---

### Minimum-phase vs zero-phase in real instruments

In this notebook, zero-phase filtering was achieved by **forward–backward filtering** (`zerophase=True`):

* excellent for **offline analysis and display**
* not usable in real time
* can create **pre-ringing** and onset smearing

In real digitizers, filtering is always **causal**.

Two important design choices are common:

* **Linear-phase (FIR) filters**

  * preserve waveform symmetry
  * introduce **larger, constant delay**

* **Minimum-phase filters**

  * minimize group delay for a given magnitude response
  * are causal and low-latency
  * introduce phase distortion

These trade-offs matter directly for:

* real-time triggering
* phase picking
* telemetry latency

We will study these in detail when we discuss **ADC, decimation, and anti-alias filtering**.

---

### Analog vs digital: what changes in practice?

#### Analog (continuous-time) intuition

* Filters described by (H(s))
* Poles and zeros in the **s-plane**
* Always causal
* Limited by hardware noise and components

#### Digital (discrete-time) reality

* Filters described by (H(z))
* Poles and zeros in the **z-plane**
* Depend on **sampling rate** and Nyquist
* Can be applied:

  * causally (one-way)
  * or acausally (two-way, offline only)

---

### Seismology-specific note: digitizers and decimation

Modern seismic digitizers typically use:

* an **anti-alias filter** before sampling
* followed by **FIR decimation filters**

Many digitizers offer choices such as:

* **linear-phase decimation**

  * more delay
  * better waveform symmetry

* **minimum-phase decimation**

  * less delay
  * more phase distortion

These choices matter for:

* real-time detection and alarms
* latency in telemetry
* consistency of arrival times across stations

We will return to this when we study:

* sampling and aliasing
* ADC
* decimation and resampling
* instrument response in the time domain

---

## Why this split works very well

After this reorganization:

* **This notebook ends with:**

  * filtering
  * onset smearing
  * convolutional model

* **The next notebook naturally begins with:**

  * sampling
  * ADC
  * anti-alias filters
  * minimum-phase vs linear-phase in real digitizers

Pedagogically, this is ideal:

1. First: what filtering does to waveforms
2. Then: how analog becomes digital
3. Then: how hardware filtering constrains everything

---

If you’d like, next I can:

* Suggest a **title and outline** for the future ADC notebook
* Help you write the **opening section on sampling & Nyquist** in the same style
* Or help you write a short **end-of-notebook summary** that explicitly points students to “next week: sampling and ADC”.

```
