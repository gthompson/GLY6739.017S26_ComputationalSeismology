## Functions

```python
def mean(values):
    return sum(values) / len(values)

print(mean([1, 2, 3, 4]))
```

## try / except

```python
def safe_int(s):
    try:
        return int(s)
    except ValueError:
        return None

print(safe_int("123"))
print(safe_int("abc"))
```

## Importing modules
