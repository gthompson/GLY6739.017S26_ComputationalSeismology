### 1. Numbers

Python is dynamically typed and assigning something to a variable will give it that type.

````python
# Three basic types of numbers
a = 1             # Integers
b = 7.5           # Floating Point Numbers
c = 3.0 + 4j      # Complex Numbers, note the use of j for the complex part


# Arithmetics work as expected.
# Upcasting from int -> float -> complex
d = a + b         # (int + float = float)
print(d)

e = c ** 2        # c to the second power, performs a complex multiplication
print(e)

# explicit type conversion
i = int(d)
print(type(i), i)    # int

# division
print(i / 3)    # float division
print(i // 3)   # integer (floor) division
print(i % 3)    # remainder after integer division

# Boolean values are a subtype of integers
t = True    # Logical 1
f = False   # Logical 0
print(t==f)      # False
print(t+4)       # 5

# There are several ways to print numbers - each involves building strings. The most useful one is f-strings.
print(f'The value of d is {d}, and the value of e is {e}')

# We can print to arbitrary precision
print(f'The value of e to 3 decimal places is {e:.3f}')
````

### 2. Strings

Just enclose something in single or double quotes and it will become a string. On Python 3 it defaults to unicode strings, e.g. non Latin alphabets and other symbols.

````python
# You can use single or double quotes to create strings.
location = "New York"

# Concatenate strings with plus.
where_am_i = 'I am in ' + location

# Or use f-strings (formatted string literals).
where_am_i = f'I am in {location}'

# Print things with the print() function.
print(location, 1, 2)
print(where_am_i)

# Strings have a lot of attached methods for common manipulations.
print(location.lower())

# Access single items with square bracket. Negative indices are from the back.
print(location[0], location[-1])

# Strings can also be sliced to get substrings.
print(location[4:])

# String can be split into lists of substrings.
words = where_am_i.split(' ')
print(words)

# We can join lists of strings back into single strings.
joined = ' '.join(words)
print(joined)

# Finally, we can replace parts of strings.
new_location = where_am_i.replace('New York', 'London')
print(new_location)
````

Strings are **immutable** - you cannot change them in place.

### 3. Lists, Tuples, and Sets

Python has 3 main collection types: Tuples, Lists, and Dictionaries. Let's start with Lists:

````python
# List use square brackets and are simple ordered collections of things.
everything = [a, b, c, 1, 2, 3, "hello", b, 2, "hello"]

# Access elements with the same slicing/indexing notation as strings.
# Note that Python indices are zero based!
print(everything[0])
print(everything[:3])
print(everything[2:-2])

# Negative indices are counted from the back of the list.
print(everything[-3:])

# Append things with the append method.
everything.append("you")
print(everything)

# There are two ways to define an empty List:
empty_list1 = []
empty_list2 = list()
````

Tuples are essentially just **immutable** lists, and we build them using round brackets (parentheses).

````python
# Tuples use round brackets and are simple ordered collections of things.
everything = (a, b, c, 1, 2, 3, "hello", b, 2, "hello")

# Access elements with the same slicing/indexing notation as Lists
# Note that Python indices are zero based!
print(everything[0])
print(everything[:3])
print(everything[2:-2])

# Negative indices are counted from the back of the list.
print(everything[-3:])

# There are two ways to define an empty Tuple:
empty_tuple1 = ()
empty_tuple2 = tuple()

# So far it behaves just like a List, but Tuples are immutable - you cannot change them.
everything.append("you")  # This line causes an error
````

Sets are also like lists, but they can only store unique values.

````python
# Sets use curly brackets and are simple ordered collection of **unique** things.
everything = {a, b, c, 1, 2, 3, "hello", b, 2, "hello"}
print(type(everything), everything)

# WARNING! Dictionaries also used curly brackets, but are key-value pairs! (See below)

# We can add new items with the add() method. (We cannot append)
everything.add("you")
print(everything)   

# Adding an existing item has no effect
everything.add("you")
print(everything)

# We can remove items with the remove() method
everything.remove(2)
print(everything)

# If we remove an item that is not present, we get an error.
# everything.remove(42)  # This line would cause an error
# So instead we can discard it
everything.discard(42)  # This line does not cause an error
print(everything)

# We can check for membership with the in keyword
3 in everything  # Returns True
42 in everything  # Returns False
# Sets are much faster than Lists for membership checking

# set theory operations
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}        
print(A.union(B))          # A or B
print(A.intersection(B))   # A and B
print(A.difference(B))     # A but not B
print(A.symmetric_difference(B))  # in A or B but not both: "Exclusive OR"

# number of elements
print(len(everything))

# copy a set to a new variable
C = everything.copy()
print(C)

# empty a set
C.clear()

# if you want to create a constant set, use frozenset
D = frozenset([1, 2, 3, 4])
print(D)    
# This is immutable - you cannot add or remove elements
# D.add(5)   # This line would cause an error

# define an empty Set:
empty_set = set()
````

Sets are often used to remove duplicates from a List or Tuple:

````python
stations = ['STA1', 'STA2', 'STA2', 'STA3', 'STA4', 
            'STA1', 'STA4', 'STA5', 'STA6']

stations_unique = list(set(stations)) # find unique values, and convert back to list
print(type(stations_unique), stations_unique)
````

We can sort lists:

````python
print(sorted(stations_unique))
````

We can loop over Tuples, Lists, and Sets:

````python
# List
print('\nlist')
for station in stations_unique:
    print(station)

# Tuple
print('\ntuple')
for station in tuple(stations_unique):
    print(station)

# Set
print('\nset')
for station in set(stations_unique):
    print(station)
````

### 4. Dictionaries

The other main collection type in Python are dictionaries. They are similiar to associative arrays or (hash) maps in other languages. Each entry is a key-value pair.

````python
# Dictionaries have named fields and no inherent order. As is
# the case with lists, they can contain anything.
information = {
    "name": "Hans",
    "surname": "Mustermann",
    "age": 78,
    "kids": [1, 2, 3]
}

# Acccess items by using the key in square brackets.
print(information["kids"])

# Add new things by just assigning to a key.
print(information)
information["music"] = "jazz"
print(information)

# Delete things by using the del operator
del information["age"]
print(information)
````

### 5. Functions

The key to conquer a big problem is to divide it into many smaller ones and tackle them one by one. This is usually achieved by using functions.

````python
# Functions are defined using the def keyword.
def do_stuff(a, b):
    return a * b

# And called with the arguments in round brackets.
print(do_stuff(2, 3))

# Python function also can have optional arguments.
def do_more_stuff(a, b, power=1):
    return (a * b) ** power

print(do_more_stuff(2, 3))
print(do_more_stuff(2, 3, power=3))

# For more complex function it is oftentimes a good idea to 
#explicitly name the arguments. This is easier to read and less error-prone.
print(do_more_stuff(a=2, b=3, power=3))
````

### 7. Control Flow

Loops and conditionals are needed for any non-trivial task. Please note that **whitespace matters in Python**. Everything that is indented at the same level is part of the same block. By far the most common loops in Python are for-each loops as shown in the following. While loops also exist but are rarely used.

````python
temp = ["a", "b", "c"]

# The typical Python loop is a for-each loop, e.g.
for item in temp:
    # Everything with the same indentation is part of the loop.
    new_item = item + " " + item
    print(new_item)
    
print("No more part of the loop.")    
````

````python
# Useful to know is the range() function.
for i in range(5):
    print(i)
````

The second crucial control flow structure are if/else conditional and they work the same as in any other language.

````python
# If/else works as expected.
age = 77

if age >= 0 and age < 10:
    print("Younger ten.")
elif age >= 10:
    print("Older than ten.")
else:
    print("wait what?")
````

````python
# List comprehensions are a nice way to write compact loops.
# Make sure you understand this as it is very common in Python.

a = list(range(10))
print(a)
b = [i for i in a if not i % 2]
print(b)

# Equivalant loop for b.
b = []
for i in a:
    if not i % 2:
        b.append(i)
print(b)
````

## The Scientific Python Ecosystem

The [SciPy Stack](https://www.scipy.org/stackspec.html) forms the basis for essentially all applications of scientific Python. Here we will quickly introduce the three core libraries:

* `NumPy`
* `SciPy`
* `Matplotlib`

The SciPy stack furthermore contains `pandas` (library for data analysis on tabular and time series data) and `sympy` (package for symbolic math), both very powerful packages, but we will omit them in this tutorial.

### 9. NumPy

Large parts of the scientific Python ecosystem use NumPy, an array computation package offering N-dimensional, typed arrays and useful functions for linear algebra, Fourier transforms, random numbers, and other basic scientific tasks.

````python
import numpy as np

# Create a large array with with 1 million samples.
x = np.linspace(start=0, stop=100, num=int(1E6), dtype=np.float64)

# Most operations work per-element.
y = x ** 2

# Uses C under the hood for speed.
print(y.sum())

# FFT and inverse
x = np.random.random(10000)
large_X = np.fft.fft(x)
x = np.fft.ifft(large_X)
````

### 10. SciPy

`SciPy`, in contrast to `NumPy` which only offers basic numerical routines, contains a lot of additional functionality needed for scientific work. Examples are solvers for basic differential equations, numeric integration and optimization, spare matrices, interpolation routines, signal processing methods, and a lot of other things.

````python
from scipy.interpolate import interp1d

x = np.linspace(0, 10, num=11, endpoint=True)
y = np.cos(-x ** 2 / 9.0)

new_x = np.linspace(0, 10, num=101, endpoint=True)

# Cubic spline interpolation to new points.
new_y = interp1d(x, y, kind="cubic")(new_x)

# Plot it (technically introduced in the next paragraph).
plt.scatter(x, y, color="red")
plt.plot(new_x, new_y);
````

### 11. Matplotlib

Plotting is done using `Matplotlib`, a package for greating high-quality static plots. It has an interface that mimics Matlab which many people are familiar with.

````python
import matplotlib.pyplot as plt

plt.plot(np.sin(np.linspace(0, 2 * np.pi, 2000)), color="green",
         label="Some Curve")
plt.legend()
plt.ylim(-1.1, 1.1)
plt.show()
````
