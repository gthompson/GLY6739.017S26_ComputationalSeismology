## A simple Event class


```python
class Event:
    def __init__(self, origin_time, magnitude):
        self.origin_time = origin_time
        self.magnitude = float(magnitude)
        self.stations = []

    def add_station(self, code):
        if code not in self.stations:
            self.stations.append(code)

    def is_significant(self, threshold=3.0):
        return self.magnitude >= threshold

    def __repr__(self):
        return f"Event(time={self.origin_time}, M={self.magnitude}, stations={self.stations})"

ev = Event("2026-01-07T16:40:00Z", 3.2)
ev.add_station("MBGH")
ev.add_station("MBBY")
print(ev)
print("significant?", ev.is_significant())
```
