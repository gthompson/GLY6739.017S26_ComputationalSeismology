## A simple class

```python
class Station:
    def __init__(self, code, elev_m=None):
        self.code = code
        self.elev_m = elev_m

    def describe(self):
        return f"Station {self.code} at {self.elev_m} m"

s = Station("MBGH", 330)
print(s.describe())
```

## Inheritance

```python
class SensorStation(Station):
    def __init__(self, code, elev_m=None, channels=None):
        super().__init__(code, elev_m)
        self.channels = channels or []

    def describe(self):
        return super().describe() + f" with {self.channels}"

ss = SensorStation("MBGH", 330, ["EHZ", "EHN", "EHE"])
print(ss.describe())
```
