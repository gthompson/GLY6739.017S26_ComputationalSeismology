## if / elif / else


```python
magnitude = 3.4

if magnitude >= 5:
    label = "large"
elif magnitude >= 3:
    label = "moderate"
else:
    label = "small"

print(label)
```

## Looping over stations and data


```python
stations = ["MBGH", "MBBY", "MBLG"]

for st in stations:
    print("Processing station:", st)
```

## Filtering with loops


```python
magnitudes = [1.2, 2.8, 3.0, 4.5, 2.1]
significant = []

for m in magnitudes:
    if m >= 3.0:
        significant.append(m)

print(significant)
```
