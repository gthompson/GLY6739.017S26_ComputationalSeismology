## if / elif / else

```python
magnitude = 3.2
if magnitude >= 5:
    print("large")
elif magnitude >= 3:
    print("moderate")
else:
    print("small")
```

## for and while loops

```python
for i in range(5):
    print(i)

x = 100
while x > 1:
    x /= 2
print("final:", x)
```
