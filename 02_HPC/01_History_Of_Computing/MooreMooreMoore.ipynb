# Moore, Moore, Moore: Moore’s Law and the Explosive Growth of Computing Power (1970–Present)

The extraordinary transformation of seismology described in the main narrative rests on a simple empirical observation known as **[Moore’s Law](chatgpt://generic-entity?number=0)**:  
> *The number of transistors on an integrated circuit roughly doubles every ~2 years.*

Although initially an observation about semiconductor manufacturing, Moore’s Law became the dominant driver of **exponential growth in computing power**, memory capacity, and storage bandwidth for nearly five decades. The machines that shaped education, research, and supercomputing can be meaningfully compared using **floating-point operations per second (FLOPS)** as a common performance metric.

This sidebar traces that growth through **representative computers from different generations**—from early home machines to modern GPUs and supercomputers.

---

## 1. A Real Home Computer (1985): BBC Micro Model B

In **1985**, my dad bought a **[BBC Micro Model B](chatgpt://generic-entity?number=1)**. This was the first computer I learned to program on. Its specifications were:

- **32 kB of RAM**
- **2 MHz CPU**
- **No hard drive**
- Programs loaded from a **cassette tape recorder**

Despite its extreme limitations by modern standards, this consumer-grade computer was already:

- **~5–10× faster than the Apollo 11 flight computer (1969)** in raw instruction speed  
  (even though the Apollo computer was far more reliable and mission-critical)

## 2. Scientific Workstations (Mid-1990s): Sun SPARCstation 2

By **1995**, during my graduate studies, I was running synthetic seismogram simulations on a **[Sun SPARCstation 2](chatgpt://generic-entity?number=2)**, which cost approximately **$15,000** at the time.

Relative to the BBC Micro, this workstation delivered roughly:

- **~1,500× faster floating-point performance**
- **~1,000× more RAM**
- A **real hard drive with fast disk I/O**

<p align="center">
  <img src="sunsparc2.png" width="500">
</p>

This leap in performance—only about **10 years later**—is a vivid illustration of Moore’s Law in action. Tasks that were impossible or impractical in the 1980s became routine by the mid-1990s.

---

## 3. My Desktop Today: Mac mini M4 (2020s Consumer Computing)

Today, I use a **[Mac mini M4](chatgpt://generic-entity?number=3)**. Compared with the Sun SPARCstation I used at the start of my PhD ~30 years ago:

- **Disk I/O:** ~**2,000× faster**  
- **RAM:** ~**1,000× larger**  
- **CPU:** ~**30,000× faster**  
- **GPU:** ~**7,000,000× faster**

And this is still a **consumer-grade desktop machine**, not a high-performance computing system. This scale of improvement over three decades would have been considered physically impossible in the 1980s.

---

## 4. The First True Supercomputer (1976): Cray-1

In **1976**, when Steve McNutt was beginning graduate school, the world’s first true supercomputer—the **[Cray-1](chatgpt://generic-entity?number=4)**—was installed at **[Los Alamos National Laboratory](chatgpt://generic-entity?number=5)**.

- Original cost: **~$8–10 million in the 1970s**  
- Inflation-adjusted: **~$40–50 million today**

<p align="center">
  <img src="seymourcray1.png" width="500">
</p>

Despite its historic status, the Cray-1 was only:

- **~10× faster** than the Sun SPARCstation I would use about **20 years later**

And remarkably:

- My **Mac mini M4 today is still**  
  - **~3,000× faster than a Cray-1 (CPU only)**  
  - **Up to ~600,000× faster using the GPU**

This is one of the most striking empirical confirmations of Moore’s Law across multiple hardware generations.

## 5. The World’s Fastest Supercomputer Today: El Capitan

Today, one of the world’s fastest supercomputers is **[El Capitan supercomputer](chatgpt://generic-entity?number=6)**, located at **[Lawrence Livermore National Laboratory](chatgpt://generic-entity?number=7)**.

<p align="center">

To match its peak performance using consumer hardware, I would need approximately:

- **~28,000 Mac mini M4 GPUs running in parallel**

This vividly illustrates the modern split between:
- **Highly specialized exascale systems** for national-scale computation  
- And **astonishingly powerful personal machines** that now rival or exceed historical supercomputers

## 6. What Moore’s Law Really Meant for Science

Over roughly **50 years**, Moore’s Law delivered:

- **~10¹²× growth in raw compute power**
- **~10⁹× growth in memory capacity**
- **~10⁸× growth in storage density**
- **~10⁶× growth in I/O bandwidth**

For science, this did not merely accelerate old workflows—it **created entirely new kinds of questions that could be asked**. In seismology, this enabled the transition from:

- Manual phase picks →  
- Triggered digital waveforms →  
- Continuous continental-scale archives →  
- And finally to **planetary-scale machine-learning-driven analysis**

The machines shown in this sidebar—BBC Micro, SPARCstation, Mac mini, Cray-1, and El Capitan—are not simply faster versions of one another. Each represents a distinct **computing regime**, with fundamentally different ideas about who could compute, what could be stored, and what could be analyzed.

They together form the quantitative backbone behind the qualitative transformation of seismology described in the main narrative.
